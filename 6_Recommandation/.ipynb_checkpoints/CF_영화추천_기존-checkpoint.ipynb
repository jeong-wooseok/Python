## 행렬분해 기반 협업필터링 추천
특정영화와 비슷한 영화를 추천해주기 위해 svd 활용
* [Singular Value Decomposition](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.TruncatedSVD.html)

```python
user_mtx = pd.pivot_table(data=ratings, values = 'rating', columns='title',index='userId')
user_mtx = user_mtx.fillna(0)
user_mtx
```

```python
item_mtx.shape
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=12) # 9062개의 영화를 12개의 요소값으로 분류
svd_mtx = svd.fit_transform(item_mtx)
svd_mtx.shape
```

```python
# 데이터 간 피어슨상관계수
corr = np.corrcoef(svd_mtx)
corr.shape

# 이 상관계수를 활용하여 '특정영화'와 상관계수 높은 영화를 뽑아주도록~
```

```python
titles = user_mtx.columns
title_list = list(titles)

idx = title_list.index("'Round Midnight (1986)")
corr_movie = corr[idx]
```

```python
corr_movie.max(),corr_movie.min(), np.quantile(corr_movie,0.9)
titles[corr_movie>np.quantile(corr_movie,0.9)][:5]
```

```python
titles = user_mtx.columns
title_list = list(titles)

def simular_movies(movie) :
    idx = title_list.index(movie)
    global corr, titles
    corr_dix = corr[idx]
    return list(titles[corr_movie>np.quantile(corr_movie,0.996)]) # 0.4% * 9064 

simular_movies("$9.99 (2008)")
```
