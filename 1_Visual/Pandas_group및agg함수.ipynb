```python
import pandas as pd
import numpy as np
df = pd.read_csv('./pd/phone_data.csv', index_col=0)
df.head(2)
```

```python
df['year'] = df.month.apply(lambda x : x.split('-')[0])
```

```python
df.head(2)
```

```python
grouped = df.groupby(['item','year','network'])
```

```python
round(grouped['duration'].agg([np.mean, np.sum]))
```

```python
#아이템별 사용시간
df.groupby('item')['duration'].sum()
```

```python
grouped = df.groupby(['item'])
```

```python
#call 사용시간

grouped.get_group('call')['duration'].sum()
```

```python
#call의 월별 사용시간
grouped = df.loc[df['item'] == 'call'].groupby('month')
```

```python
grouped['duration'].agg([np.sum,np.mean])
```

```python
#시간대별 call 사용시간은?
df
```

```python
df['time']= df['date'].apply(lambda x : x[-5:-3])
df
```

```python
grouped = df.loc[df['item'] =='call'].groupby('time')
round(grouped['duration'].agg([np.sum,np.mean]))
```

```python
grouped = df.loc[df['item'] =='call'].groupby('time')
round(grouped['duration'].agg([np.sum,np.mean])).head(3)
```

```python
grouped = df.loc[df['item'] =='call'].groupby('time')
s = grouped['duration'].sum()
s[s == s.max()]
```
