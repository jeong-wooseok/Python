```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

x = [1,2,3,4,5]
y = [6,7,8,9,10]

x1= [1,2,3,4,5]
y1 = [i**2 for i in x1]
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

x1 = range(50)
y1 = [np.cos(v) for v in x1]

x2 = range(50)
y2 = [np.sin(v) for v in x2]

plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

plt.subplot(1,2,1) # subplot : 도화지를 나눔, 1행/2열로 나누고 첫번째도화지에...
plt.plot(x1,y1, c = 'grey', ls= 'dotted',lw ='2') #x1,y1을 그리고
plt.subplot(1,2,2) # 두번째도화지에 두번째 그래프를 그려라.
plt.plot(x2,y2, color = 'black', linestyle = 'dashed', linewidth ='0.5' )
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

x1 = range(100)
y1 = [v for v in x1]


x2 = range(100)
y2 = [v*2 for v in x1]

plt.plot(x1,y1, c = 'grey', ls= ':',lw ='1',label='line1')
plt.plot(x2,y2, c = '#0000ff', ls= '-',lw ='1',label='line2')
plt.legend()

plt.text(80,100,'아무텍스트', c='darkgrey', size = '12')
plt.annotate('Check',xytext=(47,75), xy=(50,60),\
        arrowprops = dict(fc= 'lightyellow', shrink = 0.2))
# annotate : 표식 표기
# 텍스트위치 xytext(,) 표식위치 xy(,), 페이스컬러 fc, 화살표길이 shrink

plt.xlabel('M Axis',size = '15')
plt.ylabel('F Axis',size = '15')
plt.title('F = ma',size = '20')

# 더 자세한 내용은 https://matplotlib.org/에서 배울수 있음
```

### 분산그래프

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

x = np.random.rand(100)
y = x + np.random.randn(100)/5

plt.scatter(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

data1 = np.random.rand(100,2) #랜덤으로 2개 요소 지니는 100개의 리스트 Array 생성
data2 = np.random.rand(100,2)

plt.scatter(data1[:,0],data1[:,1],c='r',marker='x') # x축을 0인덱스 요소로, y축을 1인덱스 요소로 선택
plt.scatter(data2[:,0],data2[:,1],c='b',marker='o')
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

data1 = np.random.rand(50) 
data2 = np.random.rand(50)

random_nums = np.random.rand(50) # 색상도 랜덤으로 줄 수 있음 # 타입에 따라 조건문에 따라 설정 가능
area = np.pi*(20*np.random.rand(50))**2 #원의 면적 # 규모를 나타낼 때 나타내면 좋음

plt.scatter(data1,data2, c = random_nums, s = area)
```

### 막대그래프  (bar plot)

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
data = np.array([[5,25,52,30],[4,23,51,17],[6,22,52,19],[6,22,52,19]])

x = np.arange(1,9,2)
y = data[0]

plt.bar(x,y, color = '#6A8799', width=0.5) # bar차트는 단축키가 안됨
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(x,data[0],color='#6A8799',width=0.5)
plt.bar(x+0.5,data[1],color='#D59F71',width=0.5)
plt.bar(x+1,data[2],color='#F6E1D0',width=0.5)
```

```python
data[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
# data[2]:매출, data[1]:매출이익, data[0]:영업이익 x축 1~4분기로 이해 가능
plt.bar(x,data[0],color='#6A8799',width=1.2) 
plt.bar(x,data[1],color='#D59F71',width=1.2,bottom=np.sum(data[:1],axis=0))
plt.bar(x,data[2],color='#F6E1D0',width=1.2,bottom=np.sum(data[:2],axis=0))
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

woman = np.array([5,30,45,22])
man = np.array([5,25,50,31])

x = np.arange(4)
plt.barh(x,-man,color='#6A8799')
plt.barh(x,woman,color='#D59F71')
```

### 원 그래프 (PI)

```python
plt.figure(figsize=(10,5))
size = [1242,1626,458,343]
label = ['A','B','C','D']
plt.pie(size,labels=label, autopct = '%2.1f%%')#소수점 1자리
plt.legend()          
plt.show()
```

```python
plt.figure(figsize=(10,5))
size = [1242,1626,458,343]
label = ['A','B','C','D']
plt.pie(size,labels=[label[i]+': '+str(size[i]) for i in range(len(size))], autopct = '%2.1f%%')#소수점 1자리
          
plt.show()
```

```python
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} )".format(pct, absolute)
```

```python
plt.figure(figsize=(10,5))
size= [2441,2312,1031,1233]
label = ['A','B','C','D']

# plt.pie(size, labels=label, autopct='%.1f%%')
plt.pie(size, labels=label, autopct=lambda pct: func(pct, size))

plt.legend()
plt.show()
```
