## Seaborn 활용
seaborn은 matplot에서 응용된 라이브러리로 다양한 기능을 판다스와 편하게 활용가능

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15,8)
plt.rcParams['lines.linewidth'] = 1.5
```

### 분포도 : displot
총 표본집합의 수가 어느 정도의 값에 분포되어있는지를 점검 예> 상위 25%, 하위 25%

```python
# 한종류의 상수값을 array형태의 데이터로 받음 예> 대한민국 가구당소득...
x = np.random.randn(100)
sns.distplot(x, bins = 100)
```

### 산점도 : jointplot
매트릭스 상 두 변수의 점을 점검하며 상관관계를 파악할 수 있다.

```python
#sns 내장데이터셋 tip 활용
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(tips)
```

```python
# jointplot은 판다스방식의 컬럼을 x,y축으로 설정하여 산개도 출력
# 비교하고자 하는 두개의 변수값은 서로 동등함 예>가격을 올리면 구매량에 영향을 미칠까?
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# kind 를 kde 방식으로 등원형 출력 가능
sns.jointplot(x = 'total_bill', y = 'tip', data= tips, kind = 'kde')
```

### 선형회귀 : regplot, lmplot

산점도만으로는 두 지표 간 상관관계 추세를 파악하기 힘들기에 추세선을 추가한 선형회귀 활용

```python
# 예> 결제금액이 높으면 팁에 영향을 줄까?
sns.regplot('total_bill','tip', data = tips) 
```

```python
#흡연 여부에 따라 팁을 주는 정도가 다름을 확인
sns.lmplot('total_bill','tip', data = tips, size =10, hue = 'smoker') 
```

### 막대분포도 : barplot, boxplot
셀수 없는 구분(x축)에 따른 상수값(y축)의 변화를 보고자 할때 활용

```python
# 요일 별 판매액이? 어느요일에 잘팔리는데? > 일요일이 제일 잘팔림
sns.barplot(x= 'day', y= 'total_bill', data = tips, hue = 'sex') # Hue를 통해 컬럼을 2차인덱스로 unstack화
```

```python
# 분포를 더 자세히 보고 싶을 때, 주로 결제하는 고객범위는? 일일 최고/최저결제액은?
sns.boxplot(x= 'day', y= 'total_bill', data = tips, hue = 'smoker') # Hue를 통해 컬럼을 2차인덱스로 unstack화
```

### 기타 분포도

```python
# flights 데이터셋 확보
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot('month','year','passengers')
flights_pivot.head(2)
```

```python
# 판다스에서도 groupby의 unstack형태를 통해 위와같은 표 확보가능
plt.figure(figsize = (10,8))

flights_pivot['year']>1956
sns.heatmap(flights_pivot, annot=True, fmt='d')
```
