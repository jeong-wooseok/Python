### 1. 기본지표 불러오기 (KOR_ticker)

```python
import pandas as pd
import numpy as np
df = pd.read_csv('C:/Users/Administrator/Documents/data/1.KOR_ticker.csv',encoding ='euckr', index_col=0)
df.fillna(0)
df = df.apply(lambda x: x.replace('-','0'))
df['EPS'] = df['EPS'].apply(lambda x: x.replace(',',''))
df['PER'] = df['PER'].apply(lambda x: x.replace(',',''))
df['BPS'] = df['BPS'].apply(lambda x: x.replace(',',''))
df['PBR'] = df['PBR'].apply(lambda x: x.replace(',',''))
df['종목코드']= df['종목코드'].astype(str)

df['EPS']= df['EPS'].astype(float)
df['PER']= df['PER'].astype(float)
df['BPS']= df['BPS'].astype(float)
df['PBR']= df['PBR'].astype(float)
df['전일대비']=df['전일대비'].astype(float)
df['현재가.종가.']=df['현재가.종가.'].astype(float)
df = pd.DataFrame(df)
drop_columes = ['총카운트','게시물..일련번호','일자']
df = df.drop(drop_columes,axis=1)
df['ROE'] = df['EPS'] * df['PBR'] / df['현재가.종가.']
codes = []
# names = []
for j in df.iloc[:,1].values:
#     names.append(j)
    for i in df.iloc[:,0].values:
        if len(i) ==2:
            i='0000'+i
        if len(i) ==3:
            i='000'+i
        elif len(i) ==4:
            i='00'+i
        elif len(i) ==5:
            i='0'+i
        codes.append(i)
df['종목코드']=pd.Series(codes)
df.info()
```
