# 노드, 엣지 생성 및 삭제

```python
import networkx as nx
G = nx.Graph()

G.add_node(1)
G.add_nodes_from([2, 3])

#You can also add nodes along with node attributes if your container yields 2-tuples of the form (node,node_attribute_dict):
G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])

# Nodes from one graph can be incorporated into another:
H = nx.path_graph([10,15,20])
G.add_nodes_from(H)


# 노드 1개에 3개의 노드 때려박기
H1 = nx.path_graph([10,15,20])
G.add_node(H1) 


G.remove_node(2) # 노드제거

nx.draw(G, with_labels = True)
```

```python
G.add_edge(1, 2)

# 튜플 생성 후 연결도 가능 
e = (2, 3)
G.add_edge(*e)  # unpack edge tuple*

# 튜플 로 한변에 여러개 엣지 생성 (효율적)
G.add_edges_from([(1, 2), (1, 3)])

# 엣지 제거
G.remove_edge(1, 3)

nx.draw(G, with_labels = True)
```

```python
G.clear() # 그래프 제거
nx.draw(G, with_labels = True)
```

```python
G.add_edges_from([(1, 2), (1, 3)])
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')
nx.draw(G, with_labels = True)
# G.clear()            # 그래프 제거
```

## 리스트와 딕셔너리 활용, 그래프 생성 가능

```python
edgelist = [[0, 1], [1, 2], [2, 3]]
H = nx.Graph(edgelist)  
print(list(H.edges()))
nx.draw(H, with_labels = True)
# G.clear()            # 그래프 제거
```

```python
adjacency_dict = {0: [1, 2], 1: (0, 2), 2: (0, 1)}
H = nx.Graph(adjacency_dict)  # create a Graph dict mapping nodes to nbrs
list(H.edges())
nx.draw(H, with_labels = True)
```

## 두 노드 사이의 상관계수 활용하여 그래프 생성

엣지와 weight를 이중리스트 형태로 입력하여 그래프 생성 가능 

```python
FG = nx.Graph()
FG.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])
for n, nbrs in FG.adj.items():
    for nbr, eattr in nbrs.items():
        wt = eattr['weight']
        if wt < 0.5: print(f"({n}, {nbr}, {wt:.3})")
```

```python
FG = nx.Graph()
FG.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])
for (u, v, wt) in FG.edges.data('weight'):
    if wt < 0.5:
        print(f"({u}, {v}, {wt:.3})")
```

```python
nx.draw(FG, with_labels = True) 
```

## 노드 속성 Node attributes

노드는 딕셔너리 형태로 속성값을 내포할 수 있다

```python
G = nx.Graph(day="Friday")
G.graph
```

```python
G.graph['day'] = "Monday"
G.graph
```

```python
G.add_node(1, time='5pm')
G.add_nodes_from([3], time='2pm')
G.nodes[1]
```

```python
G.nodes[1]['room'] = 714
G.nodes.data()
```

```python
nx.draw(G, with_labels = True) 
# 정보를 내포하고 있는 1, 3 노드가 보인다
```

## 엣지 속성 Edge Attributes

엣지는 가중치를 속성으로 갖는다

```python
G.add_edge(1, 2, weight=4.7 )
G.add_edges_from([(3, 4), (4, 5)], color='red')
G.add_edges_from([(1, 2, {'color': 'blue'}), (2, 3, {'weight': 8})])
G[1][2]['weight'] = 4.7
G.edges[3, 4]['weight'] = 4.2
```

```python
nx.draw(G, with_labels = True) 
```

# 그래프 기호 활용하기

## 화살표

양방향

```python
# 상호 화살표
G.add_edge(1, 2)
H = nx.DiGraph(G) 
print(list(H.edges()))
nx.draw(H, with_labels = True)
# G.clear()            # 그래프 제거
```

일방향

```python
DG = nx.DiGraph()
DG.add_weighted_edges_from([(1, 2, 0.5), (3, 1, 0.75)])
DG.out_degree(1, weight='weight')
nx.draw(DG, with_labels = True)
```

```python
MG = nx.MultiGraph()
MG.add_weighted_edges_from([(1, 2, 0.5), (1, 2, 0.75), (2, 3, 0.5)])
print(dict(MG.degree(weight='weight')))
```

```python
# 가중치 최소값으로 통일
GG = nx.Graph()
for n, nbrs in MG.adjacency():
    for nbr, edict in nbrs.items():
        minvalue = min([d['weight'] for d in edict.values()])
        GG.add_edge(n, nbr, weight = minvalue)
nx.draw(GG, with_labels = True)
```
