```python
def read_file(file):
    with open(r"D:\test2\{}".format(file),"r",encoding="UTF-8") as f:
        contents = f.readlines()
    contents_del = []
    for i in contents:
         contents_del.append(i.strip())
    title = contents_del[0]
    if re.search("relatedEntities",file):
        title = "관련주제"
    else:
        title = "관련검색어"
    time = contents_del[1]
    time = re.sub('\)|\"',"",time)
    time = re.split('\(',time)
    time_title = re.sub("\:","",time[0])
    time_country = re.split(",",time[1])[1]
    time = re.split(",",time[1])[0]
    if len(time) < 22:
        time = re.findall('[0-9]{2}. [0-9]{1,2}. [0-9]{2}.',time)
        start_time = datetime.strptime(time[0],"%y. %m. %d.")
        end_time = datetime.strptime(time[1],"%y. %m. %d.")
        timedelta = str((end_time - start_time).days) + "Days"
    else:
        time = re.findall('[0-9]{2}. [0-9]{1,2}. [0-9]{2}. [ㄱ-힣]{0,2} [0-9]{0,2}:[0-9]{0,2}',time)
        start_time = datetime.strptime(re.sub("오후|오전","",time[0]),"%y. %m. %d.  %H:%M")
        end_time = datetime.strptime(re.sub("오후|오전","",time[1]),"%y. %m. %d.  %H:%M")
        timedelta = (end_time - start_time).seconds/3600
        if timedelta == 0:
            timedelta = str((end_time - start_time).days) + "Days"
        else:
            timedelta = str((end_time - start_time).seconds/3600) + "Hours"
    for i in contents_del:
        if i == "TOP":
            index_TOP = contents_del.index(i)
        if i == "RISING":
            index_RISING = contents_del.index(i)
    if "TOP" not in contents_del:
        rising_list = []
        top_list = []
    elif "RISING" not in contents_del:
        top_list = contents_del[index_TOP+1::]
        top_list = list(filter(None, top_list))
        rising_list = []
    elif "RISING" in contents_del:
        top_list = contents_del[index_TOP+1:index_RISING]
        top_list = list(filter(None, top_list))
        rising_list = contents_del[index_RISING+1::]
        rising_list = list(filter(None, rising_list))
    a = {title+time_country+" "+str(timedelta)+" "+"toprank":top_list,title+time_country+" "+str(timedelta)+" "+"risingrank":rising_list}
    return pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in a.items() ]))
```

```python
list_search = []
for i in os.listdir(r"D:\test2"):
    a = read_file(i)
    list_search.append(a)
```
