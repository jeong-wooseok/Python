### 참고자료

openpyxl로 파이썬을 통해 엑셀제어 가능<br>
lazy하게 작업 명령을 받은 후, save 할때 한번에 실행

*[openpyxl 공식홈페이지](https://openpyxl.readthedocs.io/en/stable/)<br>
*[Tutorial](https://openpyxl.readthedocs.io/en/stable/tutorial.html#create-a-workbook)

```python
from openpyxl import Workbook

wb = Workbook()
```

#### Sheet 기능

```python
# 시트 생성
ws = wb.active
ws.title = "not" # title 부여

# 줄여서 한문장으로 생성(* 추천)
ws1 = wb.create_sheet("ord") # 순서를 지정 안하면 차례대로 생성
# or
ws2 = wb.create_sheet("first", 0) # 맨앞에 생성 원할때
# or
ws3 = wb.create_sheet("new", -1) # 바로 앞의 위치로 만들기
```

```python
#시트명 확인
wb.sheetnames #sheetname method 적용
[sheet.title for sheet in wb] # title 메소드 적용
```

#### 시트수정

```python
# 시트 복사
source = wb.active
target = wb.copy_worksheet(source)
target
```
